````python
class Counter: # name of class 
    def __init__(self, start_value):
        # self refers to the new object being created
        # We're setting the 'count' attribute of this specific object
        self.count = start_value #An attribute is like a characteristic or piece of information that belongs to an object.
        print(f"Counter created with starting value: {start_value}")

    def increment(self):
        # self refers to whichever object called this method
        # We're modifying the 'count' attribute of that specific object
        self.count += 1      # self.count += value does: Takes the current value of self.count
                                #Adds value to it Assigns the result back to self.count like self.count = self.count+ value
        print(f"Count incremented to: {self.count}")

    def decrement(self):
        # Again, self refers to the specific object
        self.count -= 1
        print(f"Count decremented to: {self.count}")

    def get_count(self):
        # self lets us access the object's current count value
        return self.count

    def reset(self):
        # self lets us modify the object's attributes
        self.count = 0
        print("Counter reset to 0")

    def add_value(self, value):
        # self is used alongside other parameters
        # value is a parameter we explicitly pass
        # self is automatically passed by Python
        self.count += value
        print(f"Added{value}.Count is now:{self.count}:")

    
# Create two independent counter objects
counter1 = Counter(0)
counter2 = Counter(100)
# Output:
# Counter created with starting value: 0
# Counter created with starting value: 100
````

# __Inheritance and Code Reusability__
*  It allows you to create a new class based on an existing class, inheriting all the attributes and methods from the original class while adding new functionality or modifying existing behavior.
* Inheritance  lets you define shared attributes and methods once in a parent class, then create child classes that automatically receive that functionality.
* The child classes can then add their own unique features without duplicating the common code.
* This approach keeps your code organized, reduces duplication, and makes maintenance much easier.
* The parent class (also called the base class or superclass) contains the general, shared functionality. 
* The child classes (also called derived classes or subclasses) inherit everything from the parent and then add or customize as   needed. 
* This creates a hierarchical relationship where specialized classes build upon more general ones.



````python
# create a  parent clas
class Person:
    def __init__ (self, name, age, email):
        self.name = name
        self.age = age
        self.email =email

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Email: {self.email}")
    def update_email(self, new_email):
        self.email = new_email
        print("Email updated successfully.")


````

# __Creating a Child Class__
* To create a child class that inherits from a parent class, we specify the parent class name in parentheses after the child class name. 
* This tells Python that the new class should inherit all attributes and methods from the parent.


````python
# creating a child class
class Students(Person):
    def __init__(self, name, age,email,student_number):
        super().__init__(name, age, email ) # Call the parent class's constructor to initialize inherited attributes
        self.student_number = student_number
        self.course = []
    
    def enroll_course(self, new_course):
        self.course.append(new_course)  # Fixed: now matches the attribute name
        print(f"{self.name} enrolled in {new_course} successfully.")
   
    



````

## Let's examine what's happening here.
 * The line class Student(Person): declares that Student is a child class of Person. This means Student automatically inherits all attributes and methods from Person.

* The super().__init__(name, age, email) line is crucial. The super() function gives you access to the parent class. Here, we're calling the parent's __init__ method to handle setting up the name, age, and email attributes. This avoids duplicating that initialization code in the child class.

* After calling the parent's constructor, we add attributes that are specific to students: student_id and courses. These additional attributes exist only in Student objects, not in general Person objects.

*  Notice that student has access to both the display_info() and update_email() methods from Person, plus the enroll() method that's specific to Student. The child class has everything the parent has, plus its own additions.

````python
# Test the inheritance
students = Students("John Doe", 20, "john@university.edu", "STU001")
students.enroll_course("Python Programming")
students.update_email("alice.j@school.edu")
students.display_info()
````

###__The super() Function__
* The most common use is in constructors, where you call super().__init__() to initialize the parent's attributes before adding child-specific attributes:





````python
class Employee(Person):
    def __init__(self, name, age, email, employee_id, salary):
        super().__init__(name, age, email)  # Initialize parent attributes
        self.employee_id = employee_id
        self.salary = salary  # Add child-specific attribute

#  We can also use super() to call other parent methods from within child methods
class Manager(Employee):
    def __init__(self, name, age, email, employee_id, salary, department):
        super().__init__(name, age, email, employee_id, salary)  # Call Employee's constructor
        self.department = department
        self.team_members =[] # Initialize an empty list for team members
    def display_info(self):
        super().display_info() # call parents display_info method
        print(f"Department: {self.department}")
        # Display team members with conditional formatting
        # If team_members list has items: join them with commas
        # If team_members list is empty: display 'None'
        # ', '.join(self.team_members) - takes list items and joins them with ", " separator
        # The if/else is a ternary operator: value_if_true if condition else value_if_false
        print(f"Team Members: {', '.join(self.team_members) if self.team_members else 'None'}")
        print(f"Team size: {len(self.team_members)}")
````

## Why We Don't Use `.append()` Here

The line `', '.join(self.team_members)` is **displaying** existing data, not **adding** new data to the list.

### Key Differences:

**`.append()`** - Used for **ADDING** items TO a list:
- `self.team_members.append("John")` → Adds "John" to the team_members list
- **Purpose:** Modify the list by adding new items
- **Result:** The list grows by one item

**`.join()`** - Used for **DISPLAYING** items FROM a list:
- `', '.join(self.team_members)` → Takes existing items and creates a readable string
- **Purpose:** Convert list contents into a formatted string for display
- **Result:** Returns a string, doesn't change the original list

**Think of it this way:** `.append()` puts things IN the box, `.join()` reads what's already IN the box and makes it pretty to display.

#**When to Override Methods** 
* You should override methods when the child class needs different behavior than the parent provides. 
**Common scenarios include:**

* Specialization: The child class needs a more specific version of a general parent method. A Dog class might override an Animal class's make_sound() method to return "Bark!" instead of a generic sound.

* Additional Functionality: The child needs to do everything the parent does plus more. Use super() to call the parent method, then add the extra functionality.

* Completely Different Behavior: Sometimes the child class needs entirely different logic. In these cases, you override without calling super().


````python
# Here's an example showing different override approaches:
class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance
    def calculate_interst(self):
        return self.balance * 0.01 # 1% interest
class SavingAccount(BankAccount):
    def calculate_interest(self):
        return self.balance * 0.03 # 3% interest
class CheckingAccount(BankAccount):
    def calculate_interst(self):
        return 0 # No interest for checking accounts
 
````

***Inheritance can extend beyond two levels.***
* A child class can itself become a parent to another class, creating an inheritance hierarchy. However, you should generally keep hierarchies shallow—two or three levels at most — to avoid complexity.

````python
class Vehicle: # parent class
    def __init__(self, brand, model):# constructor method
        self.brand = brand # instance attribute
        self.model = model # instance attribute

    def start(self):# instance method 
        print(f"{self.brand} {self.model} is starting") #  method body 

class Car(Vehicle):# child class inheriting from Vehicle
    def __init__(self, brand, model, num_doors):# constructor method
        super().__init__(brand, model) # call parent constructor 
        self.num_doors = num_doors # child-specific attribute

class ElectricCar(Car):# child class inheriting from Car
    def __init__(self, brand, model, num_doors, battery_capacity):# constructor method
        super().__init__(brand, model, num_doors)# call parent constructor
        self.battery_capacity = battery_capacity # child-specific attribute

    def charge(self): # instance method
        print(f"Charging {self.brand} {self.model}") # method body
````

## How `self` Works Across Inheritance - Same Object, Shared Attributes


**What happens during object creation:**

1. **ElectricCar.__init__()** runs first
   - `self.battery_capacity = "75 kWh"` (added to the tesla object)
   - Calls `super().__init__(brand, model, num_doors)`

2. **Car.__init__()** runs next
   - `self.num_doors = 4` (added to the SAME tesla object)
   - Calls `super().__init__(brand, model)`

3. **Vehicle.__init__()** runs last  
   - `self.brand = "Tesla"` (added to the SAME tesla object)
   - `self.model = "Model 3"` (added to the SAME tesla object)

### Result: ONE Object with ALL Attributes

```python
# The tesla object now contains:

tesla.model = "Model 3"         # From Vehicle  
tesla.num_doors = 4             # From Car
tesla.battery_capacity = "75 kWh"  # From ElectricCar
```

**Key Point:** There's only **ONE** `tesla` object. The `self` parameter in all three `__init__` methods refers to this same object. Each class adds its attributes to the same object instance.

**Think of it like building a house:** Each contractor (parent class) adds their part to the SAME house, not separate houses.

**Property Decorators and Pythonic Encapsulation**

* Decorators are particularly useful in OOP because they let you add functionality to your methods without changing how users of your class interact with those methods. 

**what decorators are**
* A decorator is a special syntax in Python that modifies the behavior of functions or methods. 

* You recognize decorators by the @ symbol followed by the decorator name, placed on the line directly above a function or method definition.

**Type of decorator**

 1- **@property decorator**
* allows us to create attributes that look and behave like regular attributes but actually run methods behind the scenes.
* specifically transforms methods into attribute-like access, making your code cleaner and more intuitive.
* converts a method into a "getter" that can be accessed like an attribute. Instead of calling a method with parentheses, you access it as if it were a simple attribute.

````python
# Here's a comparison showing the difference:
# Tradintional getter method
class OldStyle:
     def __init__(self, value): # Initialize with a value
         self.__value = value  # private attribute

     def get_value(self):
         return self.__value
obj = OldStyle(10)
print(obj.get_value())  # Must use parentheses

# @Property decorator method
class NewStyle:
    def __init__(self, value): # Initialize with a value
        self.__value = value # private attribute
    @property
    def value(self):
        return self.__value
obj2 = NewStyle(20)
print(obj2.value)  # no parentheses needed
    

````

````python
# complex example
class Temperature:
    def __init__(self, celsius):
        self.__celsius = celsius
    @property
    def celsius(self):
        return self.__celsius
    @property
    def farenheit(self):
        return (self.__celsius * 9/5) + 32
    @property
    def kelvin(self):
        return self.__celsius +273.15

temp = Temperature(25)
print(temp.celsius)     # 25            
print(temp.farenheit)   # 77.0
print(temp.kelvin)      # 298.15
````

**Note**
* The fahrenheit and kelvin properties are computed on-the-fly whenever you access them. They don't exist as stored attributes — they're calculated from the celsius value each time. This is more efficient than storing multiple representations of the same data.

````python
# Building a Complete Example
class Person1:
    def __init__ (self, first_name, last_name, age):
        self.__first_name = first_name
        self.__last_name = last_name 
        self.__age = age
    @property
    def first_name(self):
        return self.__first_name
    @first_name.setter # setter method
    def first_name(self, new_first_name): # setter method
        if len(new_first_name.strip()) > 0: # validation check
        
            self.__first_name = new_first_name # set the private attribute
        else:
            raise ValueError("First name cannot be empty") # raise an error
    @property # last_name getter method
    def last_name(self): # getter method
        return self.__last_name # return the private attribute

    @last_name.setter # last_name setter method
    def last_name(self, name): # setter method
        if len(name.strip()) > 0: # validation check
            self.__last_name = name.strip() # set the private attribute
        else:
            raise ValueError("Last name cannot be empty")

    @property
    def age(self): 
        return self.__age

    @age.setter
    def age(self, value):
        if 0 <= value <= 120:
            self.__age = value
        else:
            raise ValueError("Age must be between 0 and 120")
    
    @property
    def full_name(self):
        return f"{self.__first_name} {self.__last_name}"

    @full_name.setter
    def full_name(self, name):
        parts = name.split()
        if len(parts) >= 2:
            self.__first_name = parts[0]
            self.__last_name = " ".join(parts[1:])
        else:
            raise ValueError("Full name must include first and last name")
````

**Advantages of Property Decorators**
 * Property decorators provide several significant advantages over traditional getter and setter methods.

    * **Cleaner Syntax**: Your code reads more naturally. Compare account.balance versus account.get_balance(). The property version looks cleaner and more intuitive.
    
    * **Backward Compatibility**: You can start with a simple public attribute and later convert it to a property with validation without breaking existing code:

    * **Computed Attributes**: Properties let you calculate values on demand rather than storing them, which saves memory and ensures values are always current:

    * **Lazy Initialization**: Properties can defer expensive computations until the value is actually needed:


````python
# Version 1: Simple public attribute
class User1:
    def __init__(self, email):
        self.email = email

# Version 2: Add validation without breaking code
class User1:
    def __init__(self, email):
        self.__email = email

    @property
    def email(self):
        return self.__email

    @email.setter
    def email(self, value):
        if "@" in value:
            self.__email = value
        else:
            raise ValueError("Invalid email")

#Code that uses user.email works with both versions,
#  but Version 2 adds validation protection.
````

````python
# Lazy Initialization: What is this class doing? 
# The DataAnalyzer class analyzes data,but it's designed to be efficient 
# by only doing the expensive analysis work when it's actually needed.


class DataAnalyzer: 
    def __init__(self, data): # The __init__ method
        self.__data = data # Stores the data to be analyzed
        self.__analysis = None # Starts as None (no analysis done yet)

# @property: This makes analysis behave like an attribute, 
# but it's actually a method. So you can write analyzer. analysis 
# instead of analyzer.analysis().
    @property
    def analysis(self):
       # This is the smart part:First time: If no analysis has been done (None),
       #  do it now and save the resultLater times:
       #Just return the saved result (no need to recalculate)
        if self.__analysis is None:
            # Expensive computation happens only once
            self.__analysis = self.__perform_analysis()
        return self.__analysis

    def __perform_analysis(self):
        # Complex analysis code
        return sum(self.__data) / len(self.__data)
````

````python
# Create analyzer with data
analyzer = DataAnalyzer([10, 20, 30, 40, 50])
print("Analyzer created. No analysis done yet.")

# First time accessing analysis
print("\nFirst access:")
result1 = analyzer.analysis  # Triggers calculation: (10+20+30+40+50)/5 = 30
print(f"Result 1: {result1}")
# Now self.__analysis = 30 (result is saved)

# Second time accessing analysis  
print("\nSecond access:")
result2 = analyzer.analysis  # Just returns saved result: 30
print(f"Result 2: {result2}")
# No calculation needed!

# Third time
print("\nThird access:")
result3 = analyzer.analysis  # Still just returns 30
print(f"Result 3: {result3}")

print(f"\nAll results are the same: {result1 == result2 == result3}")
````

**Property Decorators vs Traditional Methods**
* When should you use property decorators versus traditional getter/setter methods? Here are practical guidelines.

**Use property decorators when**:

* The operation is lightweight and feels like simple data access
* You want your code to read naturally as attribute access
* You're providing read-only access to computed values
* You're adding validation to what looks like a simple attribute

**Use traditional methods when**:

* The operation is expensive or time-consuming
* The method needs multiple parameters
* The method name clearly describes an action (like calculate_total() or send_email())
* You're performing a significant operation beyond just getting or setting a value


````python
# Example showing when methods are better:
class Database:
    def __init__(self):
        self.__connection = None

    # Good use of method - this is an action
    def connect(self, host, port, username, password):
        self.__connection = self.__create_connection(host, port, username, password)

    # Good use of property - simple status check
    @property
    def is_connected(self):
        return self.__connection is not None
````

**Polymorphism**

````python
# building a drawing appllication tthat need to calculate the area of 
# different shapes like circles and rectangles.
class Shape: # parent class or base class : defines the interface,
     #every shape should have an area metho
    def area(self): # method to be overridden
          return 0 # default implementation

# create a specfice shape class for rectangle
class Rectangle(Shape): # child class or derived class
    def __init__(self, width , height): # constructor method
        self.width = width # instance attribute
        self.height = height # instance attribute
    def area(self): # overriding method
        return self.width * self.height # area calculation for rectangle
    
# create a specific shape class for circle
class Circle(Shape): # child class or derived class
    def __init__(self, radius): # constructor method
        self.radius = radius  # instance attribute
    def area (self): #overriding method 
        return 3.14159 *self.radius **2 # area calulation for circle
# cereate a specific shape class for triangle
class Triangle(Shape): # child class or derived class
    def __init__(self, base, height): # constructor method
        self.base = base # instance attribute
        self.height = height # instance attribute
    def area(self): # overriding method
        return 0.5 * self.base * self.height # area calculation for triangle
 # polymorphism becomes powerful. 
 # We can write code that works with any shape without knowing the specific type:  
def print_area(shape): # shape parameter expects a Shape object
    result = shape.area() # We can call this because...
    print(f"Area: {shape.area():.2f}") # calls the area method  
````

# Understanding `shape.area()` as a Method Call

**Key Point:** `shape.area()` is a **method call** in Python, not just an attribute access.

## Method Call Components:
- **`shape`** = The object (instance)
- **`.`** = The dot operator (connects object to method)  
- **`area`** = The method name
- **`()`** = Parentheses (required to actually CALL the method)

## Why Polymorphism Works:
Since all shape classes inherit from `Shape` and override the `area()` method, we can call `shape.area()` on any shape object, and Python will automatically use the correct implementation based on the object's actual type.

````python
# Let's demonstrate that area() is indeed a method call
# Create different shape objects
rectangle = Rectangle(5, 3)
circle = Circle(4)
triangle = Triangle(6, 8)

# Each of these is a METHOD CALL - notice the parentheses ()
print("Method Calls Demonstration:")
print(f"Rectangle area: {rectangle.area():.2f}")  # Calls Rectangle's area() method
print(f"Circle area: {circle.area():.2f}")        # Calls Circle's area() method  
print(f"Triangle area: {triangle.area():.2f}")    # Calls Triangle's area() method

print("\nPolymorphism in Action:")
# The print_area function can work with ANY shape because they all have area() method
print_area(rectangle)  # Calls rectangle.area() internally
print_area(circle)     # Calls circle.area() internally
print_area(triangle)   # Calls triangle.area() internally
````

## Method vs Attribute - Important Distinction

**Method Call:** `shape.area()` with parentheses `()` 
- **Action:** Executes the area calculation  
- **Result:** Returns the computed area value
- **Purpose:** Performs an operation

**Attribute Access:** `shape.radius` without parentheses
- **Action:** Retrieves stored data
- **Result:** Returns the stored value  
- **Purpose:** Gets a piece of data

## Proof that `area()` is a Method:
```python
# You can check the type of area
print(type(circle.area))    # <class 'method'>
print(type(circle.radius))  # <class 'int'> or <class 'float'>
```

The presence of `()` is what makes it a method **call** rather than just method **reference**.

**Working with Collections of Different Types**
* One of the most practical applications of polymorphism is working with collections that contain different types of objects. 
* You can iterate through a list of mixed objects and call the same method on each, letting polymorphism handle the differences.

````python
# Building on our shapes example with a list of shapes
shapes = [Rectangle(5, 3),
           Circle(7), 
          Triangle(6, 8),
          Rectangle(2, 4),
          Circle(5)]
total_area = 0
for shape in shapes:
    total_area += shape.area() # Polymorphic method call
print(f"\nTotal area of all shapes: {total_area:.2f}")
````

**Key Components for loop:**

2. **shape** - This is the loop variable (you can name it anything)

Second iteration: shape = Circle(7)
Third iteration: shape = Triangle(6, 8), etc.
3. in keyword - Connects the loop variable to the collection

4. shapes - The collection you're iterating through (your list of shape objects)

5. : colon - Marks the end of the for loop header

6. Indented code block - Everything indented under the for loop runs once for each item

Why This Works with Polymorphism:

# Each iteration, 'shape' is a different object type:shape.area()  # But ALL shapes have an area() method!
First loop: shape is Rectangle(5,3) → calls Rectangle.area()
Second loop: shape is Circle(7) → calls Circle.area()
Third loop: shape is Triangle(6,8) → calls Triangle.area()

Alternative Names (all work the same):

The magic is that Python doesn't care what type each object is - it just calls area() on whatever object shape currently holds, and polymorphism ensures each object responds with its own implementation!
