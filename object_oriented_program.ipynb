```python
class Counter: # name of class 
    def __init__(self, start_value):
        # self refers to the new object being created
        # We're setting the 'count' attribute of this specific object
        self.count = start_value #An attribute is like a characteristic or piece of information that belongs to an object.
        print(f"Counter created with starting value: {start_value}")

    def increment(self):
        # self refers to whichever object called this method
        # We're modifying the 'count' attribute of that specific object
        self.count += 1      # self.count += value does: Takes the current value of self.count
                                #Adds value to it Assigns the result back to self.count like self.count = self.count+ value
        print(f"Count incremented to: {self.count}")

    def decrement(self):
        # Again, self refers to the specific object
        self.count -= 1
        print(f"Count decremented to: {self.count}")

    def get_count(self):
        # self lets us access the object's current count value
        return self.count

    def reset(self):
        # self lets us modify the object's attributes
        self.count = 0
        print("Counter reset to 0")

    def add_value(self, value):
        # self is used alongside other parameters
        # value is a parameter we explicitly pass
        # self is automatically passed by Python
        self.count += value
        print(f"Added{value}.Count is now:{self.count}:")

    
# Create two independent counter objects
counter1 = Counter(0)
counter2 = Counter(100)
# Output:
# Counter created with starting value: 0
# Counter created with starting value: 100
```

```python
counter1.increment()
# Output: Count incremented to: 1
```

```python
counter1.increment()
# Output: Count incremented to: 2
counter1.increment()
# Output: Count incremented to: 3
```

```python
print(f"Counter 1 value: {counter1.get_count()}")
# Output: Counter 1 value: 3

# When we call counter2.increment(), Python passes counter2 as self
counter2.increment()
# Output: Count incremented to: 101

print(f"Counter 2 value: {counter2.get_count()}")
# Output: Counter 2 value: 101

# Demonstrating methods with additional parameters
counter1.add_value(10)
# Output: Added 10. Count is now: 13

counter2.add_value(50)
# Output: Added 50. Count is now: 151

# Resetting one counter doesn't affect the other
counter1.reset()
# Output: Counter reset to 0

print(f"Counter 1: {counter1.get_count()}")  # Output: Counter 1: 0
print(f"Counter 2: {counter2.get_count()}")  # Output: Counter 2: 151
```

# Controlling Access with Methods
* Getter methods (also called accessor methods) allow controlled read access to private attributes.
  
* They return the value of a private attribute without exposing the attribute itself for direct manipulation.

```python
class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.__balance = balance
    def get_balance(self):
        return self.__balance
    
        
```

```python
#account = Variable name (the label you choose)
# = = Assignment operator (puts the object into the variable)
# BankAccount("ACC123", 1000) = Object creation (makes a new BankAccount)
account= BankAccount("ACC123", 1000)
#Method Call Syntax: account.get_balance()
# account = The object (instance) you're working with
# . = The dot operator (connects object to method)
#get_balance = The method name
# () = Parentheses (required to actually call the method)
current_balance = account.get_balance()
print(f"Balance: ${current_balance}") 
```

## Setter methods (also called mutator methods) allow controlled write access to private attributes.
* They let you modify a private attribute's value.
* But you can include validation logic to ensure only valid values are set.

```python
class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number 
        self.__balance = balance
    def get_balance(self):
        return self.__balance
    def set_balance(self, new_balance):
        if new_balance>0:
            self.__balance = new_balance
            print(f"Balance updated to ${self.__balance}") 
        else:
            print( "Error: Balance cannot be negative")
        

        
 
    
```

```python
account = BankAccount("ACC123", 1000)
account.set_balance(1500)  # Output: Balance updated to $1500
account.set_balance(-500)  # Output: Error: Balance cannot be negative

print(account.get_balance())  # Output: 1500
```

# Why Use Getters and Setters?
* Validation and Data Integrity: Setter methods let you validate data before it's stored. You can check that values are in acceptable ranges, that strings aren't empty, that numbers are positive when they should be, or any other business rule that applies.

```python
class Product:
    def __init__(self, name, price):
        self.__name = name
        self.__price = price
    def get_price(self):
        return self.__price
    def set_price(self, new_price):
        if new_price>0:
            self.__price = new_price
        else:
            print("Error: Price must be positive")
    def get_name(self):
        return self.__name
    def set_name(self, new_name):
        if len(new_name.strip()) >0:
            self.__name = new_name
        else:
            print("Erorr: Name cannot be empty")
        
    
    
```

* Maintaining Consistency: Sometimes changing one attribute requires updating others to keep the object in a valid state. Methods let you coordinate these related changes.

```python
class Reclangle:
    def __init__( self, width, height):
        self.__width = width
        self.__height = height
        self.__area = width * height
    def set_width(self, new_width):
       if new_width >0:
           self.__width = new_width
           self.__area = self.__width * self.__height
           
    def set_width(self, new_height):
       if new_height >0:
           self.__height= new_height
           self.__area = self.__width * self.__height
    def get_area(self):
        return self.__area
        
    
      
        
```

* Flexibility to Change Implementation: If you use methods to access data, you can change how data is stored internally without breaking code that uses your class. The external interface stays the same even if the internal details change.
* Code that calls get_full_name() works with both versions, even though the internal storage changed completely.


```python
# Version 1: Store full name as one attribute
class Person:
    def __init__(self, full_name): 
        self.__full_name = full_name
    def get_full_name(self):
        return self.__full_name
# Version 2: Store first and last name separately
# But keep the same interface    
class Person:
     def __init__ (self, first_name, last_name): 
         self.__first_name = first_name
         self.__last_name = last_name 

     def get_full_name(self):
         return f"{self.__first_name} {self.__last_name}"
```

* Adding Computed Properties: Sometimes what looks like an attribute is actually calculated on demand. Methods let you compute values when needed without storing them.

```python
class Circle:
    def __init__(self, radius):
        self.__radius = radius
    
    def get_radius(self):
        # area is computed udsing the formula A = πr²
        return 3.14159*self.__radius**2
    def get_circumfrence(self): 
        # circumfrence is computed using the formula C = 2πr
        return 2*3.14159*self.__radius



        
```

# Buliding a complete Example 
###  Building a more realistic class that demonstrates encapsulation principles. We'll create a Student class that manages student information with appropriate protection.

```python
class Student:
    def __init__(self, student_id, name, age):
        self.__student_id = student_id # private attribute
        self.__name = name
        self.__age = age 
        self.__grades = [] # private attribute
    def get_student_id(self):
        return self.__student_id 
    # When we want an attribute to be truly private and read-only, 
    # we don't provide a setter method. 
    # This is a key principle of encapsulation.
    def get_name(self):
        return self.__name
    def set_name(self, new_name):
        if len(new_name.strip()) > 0:
            # The strip() method removes whitespace characters (spaces, tabs, newlines) 
            # from the beginning and end of a string.
            # Without strip(), someone could set a name to just spaces,
            #  which would pass a simple length check but isn't really a valid name:
            self.__name = new_name
            return True
        else:
            return False 
    def get_age(self):
        return self.__age
    def set_age(self, new_age):
        if 0 < new_age < 120:
            self.__age = new_age
            return True
        else:
            return False
    def add_grade(self, grade):
        if 0 <= grade <= 100:
            self.__grades.append(grade)
            print(f"Grade {grade} added.")
            return True
        else:
            print("Error: Grade must be between 0 and 100.")
            return False
    def get_grades(self):
        return self.__grades.copy()
         # Return a copy of copy the grades list to prevent external modification
         # The .copy() method creates a new list with the same elements as the original,
         #  but it's a completely separate list in memory.
    def calculate_average(self): 
        if len(self.__grades) == 0:
            return 0
        else:
            return sum(self.__grades) / len(self.__grades)
 #Finally, add a method to display student information:
    def display_info(self):
        avg = self.calculate_average()
        print(f"Student ID: {self.__student_id}")
        print(f"Name: {self.__name}")
        print(f"Age: {self.__age}")
        print(f"Number of grades: {len(self.__grades)}")
        print(f"Average Grade: {avg:.2f}")

        
```

##### __why self.grades = [] appears in the __init__ method even though "grades" isn't a parameter. The key insight is that not every attribute needs to come from a parameter.__ 
__Here's why:__

Parameters vs. Initial Values
When we create a Student object, we need to provide some basic information (student_id, name, age), but a new student starts with no grades yet. So we initialize __grades as an empty list.

Think of it like this: When a student enrolls in school, they have:

A student ID (provided)
A name (provided)
An age (provided)
An empty grade book (starts empty, gets filled later)
How Do We Know to Add This?
You know to add self.__grades = [] because you're thinking ahead about what the Student class will need to do. Ask yourself:

What data will this object need to store over its lifetime?
What will start with initial values vs. what gets provided during creation?

```python
# Create a Student object
student = Student("S12345", "Alice Johnson", 20)

# Access public interface
print(student.get_name())  # Output: Alice Johnson

# Add grades through controlled method
student.add_grade(85)  # Output: Grade 85 added
student.add_grade(92)  # Output: Grade 92 added
student.add_grade(78)  # Output: Grade 78 added

# Try to add invalid grade
student.add_grade(105)  # Output: Error: Grade must be between 0 and 100

# View information
student.display_info()
# Output:
# Student ID: S12345
# Name: Alice Johnson
# Age: 20
# Number of grades: 3
# Average: 85.00
```
