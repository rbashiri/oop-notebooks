```python
from counter import Counter

# Create two independent counter objects
counter1 = Counter(0)
counter2 = Counter(100)
# Output:
# Counter created with starting value: 0
# Counter created with starting value: 100
```

```python
counter1.increment()
# Output: Count incremented to: 1
```

```python
counter1.increment()
# Output: Count incremented to: 2
counter1.increment()
# Output: Count incremented to: 3
```

```python
print(f"Counter 1 value: {counter1.get_count()}")
# Output: Counter 1 value: 3

# When we call counter2.increment(), Python passes counter2 as self
counter2.increment()
# Output: Count incremented to: 101

print(f"Counter 2 value: {counter2.get_count()}")
# Output: Counter 2 value: 101

# Demonstrating methods with additional parameters
counter1.add_value(10)
# Output: Added 10. Count is now: 13

counter2.add_value(50)
# Output: Added 50. Count is now: 151

# Resetting one counter doesn't affect the other
counter1.reset()
# Output: Counter reset to 0

print(f"Counter 1: {counter1.get_count()}")  # Output: Counter 1: 0
print(f"Counter 2: {counter2.get_count()}")  # Output: Counter 2: 151
```

```python
#class is a blueprint
# object : An instance of a class
# creating a class
class Student:
    def __init__(self, name, major):
        #Instance attributes
        self.name = name
        self.major = major
        self.grades = []
    # Instance method 
    def add_grade(self, grade):
        self.grades.append(grade) # append means add at the end

    #Instance method

    def get_average(self):
           # What condition do I want to check? Example: “Is the list empty?”
        if not self.grades:
            #What should happen if the condition is true? Example: “Return 0
            return 0
        return sum(self.grades)/ len(self.grades)
    # Instance method
    def get_info(self):
        avg = self.get_average()
        return f"{self.name}, (Major: {self.major}), Average: {avg: .2f}"




    
        
```

# Controlling Access with Methods
* Getter methods (also called accessor methods) allow controlled read access to private attributes.
  
* They return the value of a private attribute without exposing the attribute itself for direct manipulation.

```python
class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.__balance = balance
    def get_balance(self):
        return self.__balance
    
        
```

```python
#account = Variable name (the label you choose)
# = = Assignment operator (puts the object into the variable)
# BankAccount("ACC123", 1000) = Object creation (makes a new BankAccount)
account= BankAccount("ACC123", 1000)
#Method Call Syntax: account.get_balance()
# account = The object (instance) you're working with
# . = The dot operator (connects object to method)
#get_balance = The method name
# () = Parentheses (required to actually call the method)
current_balance = account.get_balance()
print(f"Balance: ${current_balance}") 
```

## Setter methods (also called mutator methods) allow controlled write access to private attributes.
* They let you modify a private attribute's value.
* But you can include validation logic to ensure only valid values are set.

```python
class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number 
        self.__balance = balance
    def get_balance(self):
        return self.__balance
    def set_balance(self, new_balance):
        if new_balance>0:
            self.__balance = new_balance
            print(f"Balance updated to ${self.__balance}") 
        else:
            print( "Error: Balance cannot be negative")
        

        
 
    
```

```python
account = BankAccount("ACC123", 1000)
account.set_balance(1500)  # Output: Balance updated to $1500
account.set_balance(-500)  # Output: Error: Balance cannot be negative

print(account.get_balance())  # Output: 1500
```

# Why Use Getters and Setters?
* Validation and Data Integrity: Setter methods let you validate data before it's stored. You can check that values are in acceptable ranges, that strings aren't empty, that numbers are positive when they should be, or any other business rule that applies.

```python
class Product:
    def __init__(self, name, price):
        self.__name = name
        self.__price = price
    def get_price(self):
        return self.__price
    def set_price(self, new_price):
        if new_price>0:
            self.__price = new_price
        else:
            print("Error: Price must be positive")
    def get_name(self):
        return self.__name
    def set_name(self, new_name):
        if len(new_name.strip()) >0:
            self.__name = new_name
        else:
            print("Erorr: Name cannot be empty")
        
    
    
```

* Maintaining Consistency: Sometimes changing one attribute requires updating others to keep the object in a valid state. Methods let you coordinate these related changes.

```python
class Reclangle:
    def __init__( self, width, height):
        self.__width = width
        self.__height = height
        self.__area = width * height
    def set_width(self, new_width):
       if new_width >0:
           self.__width = new_width
           self.__area = self.__width * self.__height
           
    def set_width(self, new_height):
       if new_height >0:
           self.__height= new_height
           self.__area = self.__width * self.__height
    def get_area(self):
        return self.__area
        
    
      
        
```

* Flexibility to Change Implementation: If you use methods to access data, you can change how data is stored internally without breaking code that uses your class. The external interface stays the same even if the internal details change.
* Code that calls get_full_name() works with both versions, even though the internal storage changed completely.


```python
# Version 1: Store full name as one attribute
class Person:
    def __init__(self, full_name): 
        self.__full_name = full_name
    def get_full_name(self):
        return self.__full_name
# Version 2: Store first and last name separately
# But keep the same interface    
class Person:
     def __init__ (self, first_name, last_name): 
         self.__first_name = first_name
         self.__last_name = last_name 

     def get_full_name(self):
         return f"{self.__first_name} {self.__last_name}"
```

* Adding Computed Properties: Sometimes what looks like an attribute is actually calculated on demand. Methods let you compute values when needed without storing them.

```python
class Circle:
    def __init__(self, radius):
        self.__radius = radius
    
    def get_radius(self):
        # area is computed udsing the formula A = πr²
        return 3.14159*self.__radius**2
    def get_circumfrence(self): 
        # circumfrence is computed using the formula C = 2πr
        return 2*3.14159*self.__radius



        
```
